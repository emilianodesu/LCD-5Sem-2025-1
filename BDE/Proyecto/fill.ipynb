```python
import psycopg2
import random
```

```python
conn = psycopg2.connect(database='proyecto_final',
                        user='bde_user',
                        host='67.217.242.23',
                        password='bd3_User#')
```

```python
cur = conn.cursor()
```

```python
consulta = """SELECT p.id_estado, p.id_municipio, p.tmax, p.tmin, p.cc, p.desciel, p.dirvienc, p.dirvieng, p.velvien, p.probprec, p.prec
FROM estado e
    JOIN municipio m ON e.id_estado = m.id_estado
    JOIN prediccion p ON m.id_estado = p.id_estado AND m.id_municipio = p.id_municipio
WHERE p.fecha = '2024-11-19'
"""
```

```python
cur.execute(consulta)
rows = cur.fetchall()
```

```python
for row in rows:
    print(row)
```

```python
prediccion = {'id_estado': [], 'id_municipio': [], 'tmax': [], 'tmin': [], 'cc': [], 'desciel': [], 'dirvienc': [], 'dirvieng': [], 'velvien': [], 'probprec': [], 'prec': []}
```

```python
for row in rows:
    prediccion['id_estado'].append(row[0])
    prediccion['id_municipio'].append(row[1])
    prediccion['tmax'].append(row[2])
    prediccion['tmin'].append(row[3])
    prediccion['cc'].append(row[4])
    prediccion['desciel'].append(row[5])
    prediccion['dirvienc'].append(row[6])
    prediccion['dirvieng'].append(row[7])
    prediccion['velvien'].append(row[8])
    prediccion['probprec'].append(row[9])
    prediccion['prec'].append(row[10])
```

```python
# Add a random integer number between -3 and 3 to tmax and tmin
prediccion['tmax'] = [tmax + random.choices([-3, -2, -1, 0, 1, 2, 3])[0] for tmax in prediccion['tmax']]
prediccion['tmin'] = [tmin + random.choices([-3, -2, -1, 0, 1, 2, 3])[0] for tmin in prediccion['tmin']]
```

```python
# Insert into database
for i in range(len(prediccion['id_estado'])):
    cur.execute(f"INSERT INTO prediccion (id_estado, id_municipio, fecha, tmax, tmin, cc, desciel, dirvienc, dirvieng, velvien, probprec, prec, ndia) VALUES ({prediccion['id_estado'][i]}, {prediccion['id_municipio'][i]}, '2024-11-25', {prediccion['tmax'][i]}, {prediccion['tmin'][i]}, {prediccion['cc'][i]}, '{prediccion['desciel'][i]}', '{prediccion['dirvienc'][i]}', {prediccion['dirvieng'][i]}, {prediccion['velvien'][i]}, {prediccion['probprec'][i]}, {prediccion['prec'][i]}, 'lunes');")
```
