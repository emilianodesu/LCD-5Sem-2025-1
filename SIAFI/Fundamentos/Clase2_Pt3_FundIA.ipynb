## **Limpieza de datos**
La limpieza de datos es un paso fundamental en el anlisis de datos y el desarrollo de modelos de inteligencia artificial. Consiste en identificar y corregir problemas en los datos que puedan afectar la calidad y precisin de los anlisis o modelos.

Es necesaria por:


* **Calidad y precisin**: Datos errneos, incompletos o duplicados pueden llevar a conclusiones incorrectas o a que los modelos de IA tengan un rendimiento deficiente.
* **Facilita el anlisis**: Al tener datos homogneos y bien estructurados, los procesos de anlisis y entrenamiento son ms eficientes y efectivos.
* **Mejora la interpretabilidad**: Datos limpios permiten que los resultados sean ms fciles de interpretar y comunicar.
* **Reduccin de sesgos**: La limpieza ayuda a eliminar sesgos que puedan estar presentes en los datos, mejorando la imparcialidad de los modelos de IA.

## **Conceptos**

### **Underfitting vs Overfitting**
#### **Underfitting**
Ocurre cuando un modelo es **demasiado simple** y no logra capturar los patrones subyacentes en los datos de entrenamiento. Esto resulta en un modelo que tiene un alto sesgo y comete errores tanto en los datos de entrenamiento como en los de prueba, mostrando un bajo rendimiento general.

Alguna de sus causas es que el modelo es demasiado bsico (ej., regresin lineal para datos no lineales), o los datos no estn correctamente procesados o no contienen suficiente informacin.

Y por ello el modelo no parende correctamente las relaciones de los datos, dando predicciones poco precias y con alto error.

#### **Overfitting**
Ocurre cuando un modelo es **demasiado complejo** y se ajusta excesivamente a los datos de entrenamiento, aprendiendo no solo los patrones sino tambin el ruido y las anomalas en los datos. Como resultado, el modelo tiene un alto grado de varianza, logrando un desempeo casi perfecto en los datos de entrenamiento, pero falla al generalizar en datos nuevos.

### **Bias y Variance**
#### **Bias (Sesgo)**
El sesgo es la tendencia del modelo a cometer errores sistemticos en sus predicciones. Un sesgo alto indica que el modelo es demasiado simple y no puede capturar las relaciones complejas en los datos, lo cual da lugar a errores constantes.

Es prcticamente el error que tiene el modelo en las predicciones, fin.

Para corregirlo se puede aumentar la complejidad del modelo, o agregando features o caractersticas (las columnas de nuestros dataframes)

#### **Variance (Varianza)**
La varianza es la sensibilidad del modelo a las pequeas variaciones en el conjunto de datos de entrenamiento. Un modelo con alta varianza es demasiado complejo y se ajusta demasiado a los datos de entrenamiento, capturando incluso el ruido o los errores aleatorios. Esto da como resultado un modelo que funciona bien en los datos de entrenamiento pero no generaliza adecuadamente, mostrando sobreajuste (overfitting).

Osea, como es de sensible el modelo a aprender patrones muy especficos de los datos de entrenamiento.

Para corregir la varianza, podemos agregar ms datos de entrenamiento (si es que se puede), simplificar el modelo o utilizar un mtodo de reduccin de caractersticas.

### **Outliers**
Los outliers (o valores atpicos) son puntos de datos que se desvan significativamente de la mayora de los datos en un conjunto. Estos valores pueden estar muy por encima o por debajo del rango esperado, y suelen ser inusuales o extremos en comparacin con otros puntos de datos.

Son las observaciones o datos que se separan del grupo D:

Y pueden deberse a los errores en la recopilacion de datos, variabilidad natural, eventos extraordinarios...

E importan mucho, puesto que pueden distorsionar nuestros resultados del anlisis y afectar la presicin de nuestros modelos, o pueden respresentar informacin importante como fraudes financieros o condiciones extremas del clime, entre otros.



### **Valores nulos (Missing values)**
Son datos que no estn disponibles o no han sido registrados en un conjunto de datos. Estos valores pueden aparecer en diferentes situaciones y pueden afectar significativamente el anlisis y la modelizacin, ya que los modelos de aprendizaje automtico y las tcnicas estadsticas generalmente requieren conjuntos de datos completos.

Puede deberse a errores en su recopilacin (ej. nuestro robot con sensores se apag por unos segundos, o hubo un apagn y no seguimos registrando ciertos datos, etc), tambin que ciertos datos no sean aplicables para todos los individuos de los datos (ej. una pregunta de cuntos ingresos tienes pero si se la haces a un nio te dice wtf bro, yo puro skibidi), etc.

Se pueden tratar de diferentes maneras, como eliminarlos (pero perdemos informacin), sustituir por valor estadstico (media, mediana,...), emplear un modelo para rellenar esos valores.

Haciendolo numericamente... (No le hagamos caso por ahora, lo dejamos para despus C:), pero explicando brevemente:

Los **percentiles** son valores que dividen un conjunto de datos en 100 partes iguales. Cada percentil representa el **porcentaje de datos** que se encuentran por debajo de ese valor.
* Ejemplo: Si el 25 percentil (P25) es 30, esto significa que el 25% de los datos son menores o iguales a 30.

Los **cuartiles** son valores que dividen un conjunto de datos ordenados en cuatro partes iguales. Hay tres cuartiles principales:
* **Primer cuartil (Q1)**: Es el percentil 25, es decir, el valor por debajo del cual se encuentra el 25% de los datos.
* **Segundo cuartil (Q2)**: Es el percentil 50, tambin conocido como la mediana, que divide el conjunto de datos en dos mitades iguales.
* **Tercer cuartil (Q3)**: Es el percentil 75, es decir, el valor por debajo del cual se encuentra el 75% de los datos.

Y en cuanto a los lmites, en este contexto de la deteccin de outliers utilizando el rango intercuartlico (IQR), los lmites son los valores que se utilizan para determinar qu datos se consideran atpicos.


AAAAHHHHH, tonses resumiento:
* **Percentiles** dividen los datos en 100 partes.
* **Cuartiles** dividen los datos en 4 partes.
* **Rango** mide la dispersin de los datos entre el mximo y mnimo.
* **Lmites** se utilizan para identificar valores atpicos en un conjunto de datos.

Ahora si, tal vez entendamos un poco lo siguiente :o

```python
#N de Outliers usando rango IQR
# Inicializamos un diccionario vaco para almacenar el nmero de outliers por columna
outliers = {}

for col in num_cols: # Pasando en cada columna de tipo numerica
    # Calculamos el primer cuartil (Q1) de la columna actual
    Q1 = np.percentile(data[col], 25) # Percentil es una funcion de np para caulcularlo en un conjunto de datos, con el 25 indicamos el primer cuartil

    # Calculamos el tercer cuartil (Q3) de la columna actual
    Q3 = np.percentile(data[col], 75)

    # Calculamos el rango intercuartlico (IQR) restando Q1 de Q3
    IQR = Q3 - Q1  # Rango intercuartlico

    # Establecemos el lmite inferior para detectar outliers
    lower_bound = Q1 - 1.5 * IQR

    # Establecemos el lmite superior para detectar outliers
    upper_bound = Q3 + 1.5 * IQR

    # Contamos el nmero de outliers en la columna actual y lo almacenamos en el diccionario
    # Suma el nmero de valores mayores que el lmite superior y menores que el lmite inferior
    outliers[col] = (data[col] > upper_bound).sum() + (data[col] < lower_bound).sum()

# Imprimimos el diccionario que contiene el nmero de outliers por columna
print(outliers)
```

Para el siguiente, recordamos cual es la desviacin estandar:

Variacin o dispersin de un conjunto de datos respecto a su media (promedio). Se utiliza para cuantificar la cantidad de incertidumbre o variabilidad en un conjunto de datos.

Lo siguiente lo hacemos en la suposicin de que los datos siguen una distribucin aproximadamente normal.

```python
#N de Outliers usando 3 std
outliers={}

for col in num_cols:
    # Calculamos la media (promedio) de la columna actual
    mean = data[col].mean()
    # Calculamos la desviacin estndar de la columna actual
    std = data[col].std()

    # Contamos el nmero de outliers en la columna actual y lo almacenamos en el diccionario
    # Un outlier lo definimos aqu como un valor que es mayor que la media ms 3 desviaciones estndar
    # o menor que la media menos 3 desviaciones estndar
    outliers[col] = (data[col] > (mean + 3 * std)).sum() + (data[col] < (mean - 3 * std)).sum()

print(outliers)
```

**SimpleImputer viene de sklearn**

Se utiliza para manejar los valores faltantes en conjuntos de datos. Proporciona una forma sencilla de imputar (rellenar) los datos faltantes mediante diferentes estrategias

En ese caso queremos rellenar con la mediana. Viendo la grafica de arriba sigue una distribucion bastante normal, entonces nos da igual utilizar la mediana o la media.

La diferencia es que si hay outliers, la media no nos da un valor preciso puesto que se ve afectado muchote por ese outlier

Entonces generalmente se acostumbra rellenar con la mediana.


```python
# Creamos un objeto SimpleImputer con la estrategia de imputacin como 'median' (mediana :B)
impute = SimpleImputer(strategy='median')

# Imprimimos la mediana de la columna 'Age' antes de la imputacin
print('Mediana inicial: ', data['Age'].median())

# Ajustar el imputador a los datos de la columna 'Age' y transformar (imputar) los valores faltantes
salida = impute.fit_transform(data[['Age']])

# Convertir la salida imputada en un DataFrame de pandas
salida = pd.DataFrame(salida)

# Imprimir el nmero de valores nulos en los nuevos datos imputados
print('\nNulos en los nuevos datos: ', salida.isna().sum())

# Imprimir la mediana de los nuevos datos imputados
print('\nMediana nuevos datos: ', salida.median())
```
