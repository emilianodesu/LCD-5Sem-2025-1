## Regresión lineal
---






La regresión lineal es un modelo matemático que describe la relación entre varias variables. Los modelos de regresión lineal son un procedimiento estadístico que ayuda a predecir el futuro. Se utiliza en los campos científicos y en los negocios, y en las últimas décadas se ha utilizado en el aprendizaje automático (aprendiendo machin [Machine larning]).



La tarea de la regresión en el aprendizaje automático consiste en predecir un parámetro (Y) a partir de un parámetro conocido X.

La idea clave es que existe una relación lineal entre las variables independientes (características o features) y la variable dependiente (lo que queremos predecir).

Para una regresión lineal simple con una variable independiente (X) y una variable dependiente (Y):

> **Variable independiente (X)** (o características): Características que queremos usar para predecir algún valor dado de "Y"

> **Variable dependiente (Y)** (u objetivo): Es lo que tratamos de predecir >:)

formula-regresion-lineal.webp

Los modelos de regresión lineal son muy populares en diversos campos de investigación gracias a su rapidez y facilidad de interpretación.

Debido a su capacidad para transformar datos, pueden utilizarse para simular una amplia gama de relaciones, y debido a su forma, que es más simple que la de las redes neuronales, sus parámetros estadísticos se analizan y comparan con facilidad, lo que permite que se les extraiga información valiosa.

La regresión lineal no sólo se utiliza con fines de predicción: también ha demostrado su eficacia para describir sistemas.

> Ojo ojito, las variables deben de seguir lo siguiente para que sea correcta la aplicación de este modelo de predicción:
>> 1. Debe existir una **relación lineal** entre las variables dependientes e independientes.
>>> La regresión lineal supone que hay una relación recta entre la variable que queremos predecir (la dependiente) y las variables que usamos para hacer la predicción (las independientes). Es decir, al graficar los datos, deberían poder unirse más o menos en línea recta.
>> 2. Las variables de entrada y salida son **datos limpios**.
>>> Para obtener buenos resultados, los datos que alimentas al modelo deben estar libres de errores, valores extremos (outliers), o datos faltantes. Si los datos son "sucios", el modelo no funcionará bien
>> 3. El modelo se ajustará a los datos cuando tenga **variables de entrada altamente correlacionada**
>>> La regresión lineal funciona mejor cuando las variables independientes (de entrada) están relacionadas con la variable dependiente (de salida). La "correlación" mide qué tan bien dos variables están relacionadas, y si esa relación es fuerte, el modelo se ajustará mejor. (recuerden la clase anterior :B)
>> 4. El modelo hará predicciones más confiables si sus variables tienen una distribución normal
>>> El modelo asume que los errores (la diferencia entre las predicciones y los valores reales) se distribuyen normalmente. Esto permite hacer inferencias y calcular intervalos de confianza de manera más precisa.

### **Tipos de regresión lineal**

#### **Simple**

En una regresión lineal, se trata de establecer una relación entre una variable independiente y su correspondiente variable dependiente. Esta relación se expresa como una línea recta. No es posible trazar una línea recta que pase por todos los puntos de un gráfico si estos se encuentran ordenados de manera caótica. Por lo tanto, sólo se determina la ubicación óptima de esta línea mediante una regresión lineal. Algunos puntos seguirán distanciados de la recta, pero esta distancia debe ser mínima. El cálculo de la distancia mínima de la recta a cada punto se denomina **función de pérdida**.

### **Ejemplito sencillo**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

dataset_alturas_pesos = {
    'Altura': [ 1.47 , 1.5 , 1.55, 1.6 , 1.63 , 1.68, 1.73, 1.75 , 1.78 , 1.81],
    'Peso': [ 45, 53, 55, 61, 59, 66 , 70, 75, 76, 83 ]
}
df = pd.DataFrame(dataset_alturas_pesos)
df
```

```python
df.plot.scatter(x='Altura', y='Peso')
```

```python
reg = LinearRegression()
# Entrenando con: X (datos a entrenar) y Y (datos a predecir)
reg.fit(df[['Altura']], df['Peso'])
df[['Altura']]
```

**Ojito**: los corchetes dobres de `df[['Altura']]` , se usan para que **Altura** sea tratada como un **DataaFrame** con múltiples columnas en lugar de como una **Serie**.

Es necesario porque el método `fit()` de `LinearRegression()` espera que la variable independiente (en este caso, `Altura`) sea un array de 2 dimensiones (matriz), donde cada fila es una observación y cada columna es una característica.

Ahora si ya podmeos predecir C:


```python
reg.predict([[1.90], [1.40], [1.55]])
```

Donde nosotros podemos ver la ordenada al origen con .intercept_ y su pendiente con .coef_

```python
print(f"Ordenada al origen: {reg.intercept_}")
print(f"Pendiente: {reg.coef_}")
```

Con esto entonces, podríamos calcular con la ecuación de la recta todos los valores de Y :D

```python
miau = reg.coef_ * df['Altura'] + reg.intercept_
miau
```

```python
df["Prediccion"] = miau
df
```

```python
ax = df.plot.line(x='Altura', y='Prediccion', color='red') # Guardando los ejes
df.plot.scatter(x='Altura', y='Peso', ax = ax)
```

El primer gráfico se almacena en ax, que es un objeto de tipo `Axes`, osea que contiente toda la informoacoón sobre los ejes del gráfico.
En la segunda línea se genera un gráfico de dispersión pero le pasamos el parámetro `ax=ax` donde le decimos a pandas que debe dibujar el gráfico de dispersión en los mismos ejes que ya fueron creados anteriormente.
OSEASE KE: En lugar de crear un segundo gráfico con sus propios ejes, los puntos de dispersión se agregan sobre el gráfico de línea que ya existe. Combinando ambos en uno solo :D

### Otro ejemplo: preparemos nuestros datos
---

Si entrenamos y probamos nuestro modelo en el mismo conjunto de datos, es probable que el modelo se ajuste muy bien a esos datos (**sobreajuste** u **overfitting**). Sin embargo, puede que no generalice bien para datos nuevos que nunca ha visto.

Dividir los datos permite **evaluar el rendimiento en datos** que el modelo no ha utilizado durante el entrenamiento, lo que da una estimación más realista de su capacidad para hacer predicciones sobre datos nuevos.

#### Training set

Este conjunto de datos se utiliza para **entrenar** al modelo, es decir, el modelo ajusta sus parámetros o aprende patrones a partir de estos datos.

Normalmente, representa entre el **70%** y el **80%** de los datos totales

statistical_modeling_methodology-h.png

#### Test set

Este conjunto de datos se utiliza para evaluar el modelo después del entrenamiento. El modelo nunca ha visto estos datos, por lo que la evaluación de rendimiento en este conjunto nos dice qué tan bien puede predecir sobre datos nuevos.

Normalmente, representa entre el 20% y el 30% de los datos totales.

#### wtf bro?

Osease si pero como?

La librería **scikit-learn** tiene una función muy útil llamada `train_test_split` que hace esto de forma automática y aleatoria.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Supongamos que tenemos nuestras variables independientes o características (X) y dependientes o que queremos predecir(Y)
X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([1.2, 1.9, 3.1, 3.9, 5.1])

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
print(f"X_train: {X_train}")
print(f"X_test: {X_test}")
print(f"Y_train: {Y_train}")
print(f"Y_test: {Y_test}")
```

En el código de arriba, ponemos que:
- `test_size=0.2` significa que el 20% de los datos se destina al conjunto de prueba, y el 80% al conjunto de entrenamiento.
- `random_state=0` para garantizar que la división sea reproducible. Osease que cada que le demos click en ejecutar, siempre nos aletorice igual.

En cuanto a las salidas tendríamos que:
`X_train`, `Y_train`: Son los subconjuntos de los datos de entrenamiento.
`X_test`, `Y_test`: Son los subconjuntos de los datos de prueba.

En este caso, los subconjuntos son porciones específicas de los datos originales que se dividen en grupos más pequeños, destinados para diferentes propósitos en el proceso de entrenamiento y evaluación del modelo

### Entrenemos nuestro modelo! :D

```python
from sklearn.linear_model import LinearRegression

# Crear el modelo de regresión lineal
model = LinearRegression() # Esta funcion crea un objeto modelo de regresion lineal

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, Y_train) # Ajusta el modelo a los datos de entrenamiento, encontrando los mejores valores para los coeficientes e intercepto de la ecuación de la regresión.
```

Ya que tenemos nuestro modelito, hagamos unas predicciones sencillas.

```python
# Hacer predicciones con los datos de prueba
Y_pred = model.predict(X_test)
Y_pred, Y_test
```

Grafiquemos los resultados :)

```python
ax = plt.axes()
ax.scatter(X_train, Y_train, color='blue', label='Datos de entrenamiento')
ax.scatter(X_test, Y_pred, color='red', label='Datos de prueba')
ax.legend()
```

### Evaluaciones

A chido, y ahora?
Después de hacer las predicciones, es importante evaluar qué tan bien el modelo predice los valores reales (`Y_test`). Para ello, se utilizan métricas de evaluación. Dos de las métricas más comunes en la regresión lineal son:

- **MSE (Error Cuadrático Medio)**: Mide el promedio de los errores al cuadrado entre los valores predichos y los reales. Un MSE más bajo indica mejores predicciones.
> Qué tan lejos están los valores predichos de los valores reales. Sigamos lo siguiente que son los pasos de como se calcula el MSE para entender de una manera mas simple :D :
>>1. Calculamos el error, que es simplemente la diferencia entre el valor real (lo que pasó en la vida real) y el valor predicho (lo que nuestro modelo predijo).
>>2. Para evitar que los errores negativos y positivos se cancelen entre sí (como en el caso del -3 y el 5), elevamos al cuadrado cada error.
>>3. Una vez que tenemos todos los errores al cuadrado, calculamos el promedio de estos errores.
>>4. Festejamos. Ya tenemos el MSE que es el promedio de cuánto se están equivocando nuestra spredicciones
>> Si tenemos el MSE bajo significa que los errores entre los valores predichos y los reales son pequeños, osea todo funciona de maravilla.



- **R^2 (Coeficiente de Determinación)**: Indica qué proporción de la varianza en la variable dependiente es explicada por las variables independientes. Un valor más cercano a 1 significa un mejor ajuste.
>> Dice qué porcentaje de la variabilidad en los datos reales puede ser explicado por tu modelo.
    - **Variabilidad total**: En cualquier conjunto de datos, existe una cantidad de variabilidad natural. Por ejemplo, si tenemos una lista de alturas y pesos, no todos van a tener el mismo peso, incluso si tienen la misma altura. Esa diferencia entre los datos reales y el promedio se llama **variabilidad**.
    - **Variabilidad explicada**: Un modelo de regresión intenta explicar esa variabilidad. El R² nos dice qué tan bien lo está haciendo. Si nusetro modelo es bueno, debería poder explicar gran parte de esa variabilidad.

- **R² = 1**: El modelo es perfecto y explica toda la variabilidad.
- **R² cercano a 1**: El modelo explica bien los datos, y hace predicciones cercanas a los valores reales.
- **R² cercano a 0**: El modelo no está explicando bien los datos, y hay mucha variabilidad que no puede predecir.
- **R² negativo**: Esto puede ocurrir si nuestro modelo es peor que simplemente predecir el promedio de los datos.

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calcular el MSE y el R^2
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

# Mostrar los resultados
print(f"MSE: {mse}")
print(f"R^2: {r2}")
```

Ese errorsillo de R^2 sucede porque (coeficiente de determinación) no está bien definida cuando tenemos menos de dos muestras en el conjunto de prueba.

Mide qué tan bien las predicciones se ajustan a los valores reales, pero necesita al menos dos puntos de datos para calcular una medida significativa. Con un solo punto de prueba, no se puede calcular la varianza necesaria para definirlo.

Para arreglarlo podemos aumentar el tamaño del conjunto de prueba.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
Y_pred = model.predict(X_test)
```

Ahora el conjunto de prueba será del 30% en lugar del 20%, lo que podría darnos mas ejemplos para calcular R^2 :D

O si llegasemos a trabajar con muy pocos datos, pues mejor omitir R^2.

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculamos el MSE
mse = mean_squared_error(Y_test, Y_pred)

# Verificamos si hay suficientes datos para calcular R^2
if len(Y_test) > 1:
    r2 = r2_score(Y_test, Y_pred)
    print(f"R²: {r2}")
else:
    print("No se puede calcular R² con menos de dos muestras.")

# Mostramos MSE
print(f"MSE: {mse}")
```

## Regresión logística
---

Algoritmo para predecir la **probabilidad de una clase o evento**.

A diferencia de la regresión lineal, que predice valores continuos, la regresión logística se utiliza principalmente para problemas de clasificación binaria (es decir, cuando la variable dependiente tiene dos posibles valores, como "sí" o "no", "aprobado" o "rechazado", etc.).

#### Cosas a tomar en cuanta

En regresión logística, la variable dependiente es binaria o categórica. El valor de esta variable generalmente se codifica como 0 o 1.

**Ejemplo**: En un problema de clasificación de correos electrónicos como "spam" o "no spam", la variable dependiente sería 1 para spam y 0 para no spam.

**Función sigmoide (logística)**

La regresión logística utiliza una función sigmoide para convertir la salida de una combinación lineal de variables en una probabilidad.

- Si el valor resultante de la función sigmoide es mayor que 0.5, el modelo predice 1 (la clase positiva), y si es menor que 0.5, predice 0 (la clase negativa).
- Esto puede ajustarse dependiendo del problema y el umbral que se elija.

imagen.png

#### ¿Diferencias con la regresión lineal?

La **regresión lineal** nos da una salida continua
> Por ejemplo para conocer el porcentaje de probabilidad de que llueva o el precio de una acción

Por otra parte, la **regresión logística** proporciona una salida discreta, donde podemos conocer si va a llover o no, o si el precio de una acción subirá o no.

#### **Tipos de regresión logística**



##### Binaria

- Se utiliza para predecir dos clases (0 o 1).
- Ejemplo: Clasificación de enfermedades (enfermo o no enfermo).

##### Multinomial

- Se utiliza para predecir más de dos clases sin un orden inherente.
- **Ejemplo**: Clasificación de tipos de frutas (manzana, naranja, plátano).

##### Ordinal

- Se utiliza para predecir variables con múltiples clases que tienen un orden.
- **Ejemplo**: Clasificación de la satisfacción del cliente (muy insatisfecho, insatisfecho, neutral, satisfecho, muy satisfecho).

#### Coeficientes

En la regresión logística, los coeficientes representan el cambio en el logaritmo de la razón de probabilidades (log-odds) de la clase positiva frente a la clase negativa por cada unidad de cambio en las variables independientes.

### Ejemplitos

#### Ejemplo 1

```python
from sklearn.linear_model import LogisticRegression

X = np.array([[50], [60], [70], [80], [90]])  # Calificaciones de estudio
y = np.array([0, 0, 1, 1, 1])  # Resultados (0 = no pasa, 1 = pasa)

model = LogisticRegression()
model.fit(X, y)

# Predicción
X_test = np.linspace(40, 100, 200).reshape(-1, 1) # Serie de 200 valores entre 40 y 100, osea el modelo logistico
# con el reshape hacemos el array de valores en una matriz de una sola columna. El modelo
# de regresion logistica espera varias filas y una columna
y_pred = model.predict_proba(X_test)[:, 1] # calcula las probabilidades que el alumno pase
# El [:, 1] nos selecciona la segunda columna osea la probabilidad de que el resultado sea 1

# Visualización
plt.scatter(X, y, color='red', label='Datos')
plt.plot(X_test, y_pred, color='blue', label='Modelo Logístico')
plt.axhline(0.5, color='green', linestyle='--', label='Umbral (0.5)')
plt.xlabel('Calificación de Estudio')
plt.ylabel('Probabilidad de Pasar')
plt.title('Regresión Logística: Estudiante Pasa el Examen')
plt.legend()
plt.show()
```

```python
Y_pred = model.predict(X_test)
print(f"Los datos de test: {Y_test}")
print(f"Los datos de pred: {Y_pred}")
```
