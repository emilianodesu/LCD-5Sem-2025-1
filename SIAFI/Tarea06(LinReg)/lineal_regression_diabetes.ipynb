#### **Descripción**
Realiza los siguientes pasos usando el dataset `diabetes` de sklearn, aplica el modelo de regresión lineal:
- Carga el dataset incluido en `sklearn`
- Selecciona una característica, como por ejemplo el BMI
- Divide los datos en datos de testing y entrenamiento.
- Entrena el modelito
- Evalúa el modelo usando el error cuadrático medio.

```python
from sklearn import datasets
dataset = datasets.load_diabetes()
```

```python
dataset.keys()
```

```python
print(dataset.DESCR)
```

```python
X = dataset.data
y = dataset.target
```

```python
# Plotting the data
for i in range(X.shape[1]):
    plt.figure(figsize=(5,5))
    plt.scatter(X[:,i], y)
    plt.xlabel(dataset.feature_names[i])
    plt.ylabel('Diabetes Progression')
    plt.show()
```

Los datos ya están estandarizados

```python
# Tomamos bmi como variable independiente
X = X[:,2]
X = X.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_hat = model.predict(X_test)
# Plotting the data
plt.figure(figsize=(9,6))
plt.scatter(X_test, y_test)
plt.plot(X_test, y_hat, color='red')
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.show()
```

```python
# Evaluating the model
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_hat)
print(f'MSE: {mse}')
```
