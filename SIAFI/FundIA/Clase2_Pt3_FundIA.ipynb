## **Limpieza de datos**
La limpieza de datos es un paso fundamental en el análisis de datos y el desarrollo de modelos de inteligencia artificial. Consiste en identificar y corregir problemas en los datos que puedan afectar la calidad y precisión de los análisis o modelos.

Es necesaria por:


* **Calidad y precisión**: Datos erróneos, incompletos o duplicados pueden llevar a conclusiones incorrectas o a que los modelos de IA tengan un rendimiento deficiente.
* **Facilita el análisis**: Al tener datos homogéneos y bien estructurados, los procesos de análisis y entrenamiento son más eficientes y efectivos.
* **Mejora la interpretabilidad**: Datos limpios permiten que los resultados sean más fáciles de interpretar y comunicar.
* **Reducción de sesgos**: La limpieza ayuda a eliminar sesgos que puedan estar presentes en los datos, mejorando la imparcialidad de los modelos de IA.

Nosotros nos vamos a enfocar en eliminar valores duplicados, manejar valores nulos, corregir errores, normalizar, estandarizar, filtrar...
* **Eliminación de valores duplicados**: Borrar datos que se encuentran repetidos para evitar que tengan peso innecesario.
* **Manejo de valores nulos**: Completar o eliminar registros incompletos para mantener la consistencia de los datos.
* **Corrección de errores**: Revisar y corregir errores tipográficos, formatos inconsistentes o valores fuera de rango.
* **Normalización y estandarización**: Convertir los datos a un formato uniforme por ejemplo, fechas en un mismo formato o nombres en una misma estructura.
* **Filtrado de datos irrelevantes**: Quitar información que no aporta valor o que podría introducir ruido en el análisis.

## **Conceptos**

### **Underfitting vs Overfitting**
#### **Underfitting**
Ocurre cuando un modelo es **demasiado simple** y no logra capturar los patrones subyacentes en los datos de entrenamiento. Esto resulta en un modelo que tiene un alto sesgo y comete errores tanto en los datos de entrenamiento como en los de prueba, mostrando un bajo rendimiento general.

Alguna de sus causas es que el modelo es demasiado básico (ej., regresión lineal para datos no lineales), o los datos no están correctamente procesados o no contienen suficiente información.

Y por ello el modelo no aprende correctamente las relaciones de los datos, dando predicciones poco precias y con alto error.

#### **Overfitting**
Ocurre cuando un modelo es **demasiado complejo** y se ajusta excesivamente a los datos de entrenamiento, aprendiendo no solo los patrones sino también el ruido y las anomalías en los datos. Como resultado, el modelo tiene un alto grado de varianza, logrando un desempeño casi perfecto en los datos de entrenamiento, pero falla al generalizar en datos nuevos.

### **Bias y Variance**
#### **Bias (Sesgo)**
El sesgo es la tendencia del modelo a cometer errores sistemáticos en sus predicciones. Un sesgo alto indica que el modelo es demasiado simple y no puede capturar las relaciones complejas en los datos, lo cual da lugar a errores constantes.

Es prácticamente el error que tiene el modelo en las predicciones, fin.

Para corregirlo se puede aumentar la complejidad del modelo, o agregando features o características (las columnas de nuestros dataframes)

#### **Variance (Varianza)**
La varianza es la sensibilidad del modelo a las pequeñas variaciones en el conjunto de datos de entrenamiento. Un modelo con alta varianza es demasiado complejo y se ajusta demasiado a los datos de entrenamiento, capturando incluso el ruido o los errores aleatorios. Esto da como resultado un modelo que funciona bien en los datos de entrenamiento pero no generaliza adecuadamente, mostrando sobreajuste (overfitting).

Osea, como es de sensible el modelo a aprender patrones muy específicos de los datos de entrenamiento.

Para corregir la varianza, podemos agregar más datos de entrenamiento (si es que se puede), simplificar el modelo o utilizar un método de reducción de características.

### **Outliers**
Los outliers (o valores atípicos) son puntos de datos que se desvían significativamente de la mayoría de los datos en un conjunto. Estos valores pueden estar muy por encima o por debajo del rango esperado, y suelen ser inusuales o extremos en comparación con otros puntos de datos.

Son las observaciones o datos que se separan del grupo D:

Y pueden deberse a los errores en la recopilacion de datos, variabilidad natural, eventos extraordinarios...

E importan mucho, puesto que pueden distorsionar nuestros resultados del análisis y afectar la presición de nuestros modelos, o pueden respresentar información importante como fraudes financieros o condiciones extremas del clime, entre otros.



### **Valores nulos (Missing values)**
Son datos que no están disponibles o no han sido registrados en un conjunto de datos. Estos valores pueden aparecer en diferentes situaciones y pueden afectar significativamente el análisis y la modelización, ya que los modelos de aprendizaje automático y las técnicas estadísticas generalmente requieren conjuntos de datos completos.

Puede deberse a errores en su recopilación (ej. nuestro robot con sensores se apagó por unos segundos, o hubo un apagón y no seguimos registrando ciertos datos, etc), también que ciertos datos no sean aplicables para todos los individuos de los datos (ej. una pregunta de cuántos ingresos tienes pero si se la haces a un niño te dice wtf bro, yo puro skibidi), etc.

Se pueden tratar de diferentes maneras, como eliminarlos (pero perdemos información), sustituir por valor estadístico (media, mediana,...), emplear un modelo para rellenar esos valores.

Haciendolo numericamente... (No le hagamos caso por ahora, lo dejamos para después C:), pero explicando brevemente:

Los **percentiles** son valores que dividen un conjunto de datos en 100 partes iguales. Cada percentil representa el **porcentaje de datos** que se encuentran por debajo de ese valor.
* Ejemplo: Si el 25º percentil (P25) es 30, esto significa que el 25% de los datos son menores o iguales a 30.

Los **cuartiles** son valores que dividen un conjunto de datos ordenados en cuatro partes iguales. Hay tres cuartiles principales:
* **Primer cuartil (Q1)**: Es el percentil 25, es decir, el valor por debajo del cual se encuentra el 25% de los datos.
* **Segundo cuartil (Q2)**: Es el percentil 50, también conocido como la mediana, que divide el conjunto de datos en dos mitades iguales.
* **Tercer cuartil (Q3)**: Es el percentil 75, es decir, el valor por debajo del cual se encuentra el 75% de los datos.

Y en cuanto a los límites, en este contexto de la detección de outliers utilizando el rango intercuartílico (IQR), los límites son los valores que se utilizan para determinar qué datos se consideran atípicos.


AAAAHHHHH, tonses resumiento:
* **Percentiles** dividen los datos en 100 partes.
* **Cuartiles** dividen los datos en 4 partes.
* **Rango** mide la dispersión de los datos entre el máximo y mínimo.
* **Límites** se utilizan para identificar valores atípicos en un conjunto de datos.

Ahora si, tal vez entendamos un poco lo siguiente :o

```python
#Nº de Outliers usando rango IQR
# Inicializamos un diccionario vacío para almacenar el número de outliers por columna
outliers = {}

for col in num_cols: # Pasando en cada columna de tipo numerica
    # Calculamos el primer cuartil (Q1) de la columna actual
    Q1 = np.percentile(data[col], 25) # Percentil es una funcion de np para caulcularlo en un conjunto de datos, con el 25 indicamos el primer cuartil

    # Calculamos el tercer cuartil (Q3) de la columna actual
    Q3 = np.percentile(data[col], 75)

    # Calculamos el rango intercuartílico (IQR) restando Q1 de Q3
    IQR = Q3 - Q1  # Rango intercuartílico

    # Establecemos el límite inferior para detectar outliers
    lower_bound = Q1 - 1.5 * IQR

    # Establecemos el límite superior para detectar outliers
    upper_bound = Q3 + 1.5 * IQR

    # Contamos el número de outliers en la columna actual y lo almacenamos en el diccionario
    # Suma el número de valores mayores que el límite superior y menores que el límite inferior
    outliers[col] = (data[col] > upper_bound).sum() + (data[col] < lower_bound).sum()

# Imprimimos el diccionario que contiene el número de outliers por columna
print(outliers)
```

Para el siguiente, recordamos cual es la desviación estandar:

Variación o dispersión de un conjunto de datos respecto a su media (promedio). Se utiliza para cuantificar la cantidad de incertidumbre o variabilidad en un conjunto de datos.

Lo siguiente lo hacemos en la suposición de que los datos siguen una distribución aproximadamente normal.

```python
#Nº de Outliers usando 3 std
outliers={}

for col in num_cols:
    # Calculamos la media (promedio) de la columna actual
    mean = data[col].mean()
    # Calculamos la desviación estándar de la columna actual
    std = data[col].std()

    # Contamos el número de outliers en la columna actual y lo almacenamos en el diccionario
    # Un outlier lo definimos aquí como un valor que es mayor que la media más 3 desviaciones estándar
    # o menor que la media menos 3 desviaciones estándar
    outliers[col] = (data[col] > (mean + 3 * std)).sum() + (data[col] < (mean - 3 * std)).sum()

print(outliers)
```

**SimpleImputer viene de sklearn**

Se utiliza para manejar los valores faltantes en conjuntos de datos. Proporciona una forma sencilla de imputar (rellenar) los datos faltantes mediante diferentes estrategias

En ese caso queremos rellenar con la mediana. Viendo la grafica de arriba sigue una distribucion bastante normal, entonces nos da igual utilizar la mediana o la media.

La diferencia es que si hay outliers, la media no nos da un valor preciso puesto que se ve afectado muchote por ese outlier

Entonces generalmente se acostumbra rellenar con la mediana.


```python
# Creamos un objeto SimpleImputer con la estrategia de imputación como 'median' (mediana :B)
impute = SimpleImputer(strategy='median')

# Imprimimos la mediana de la columna 'Age' antes de la imputación
print('Mediana inicial: ', data['Age'].median())

# Ajustar el imputador a los datos de la columna 'Age' y transformar (imputar) los valores faltantes
salida = impute.fit_transform(data[['Age']])

# Convertir la salida imputada en un DataFrame de pandas
salida = pd.DataFrame(salida)

# Imprimir el número de valores nulos en los nuevos datos imputados
print('\nNulos en los nuevos datos: ', salida.isna().sum())

# Imprimir la mediana de los nuevos datos imputados
print('\nMediana nuevos datos: ', salida.median())
```
