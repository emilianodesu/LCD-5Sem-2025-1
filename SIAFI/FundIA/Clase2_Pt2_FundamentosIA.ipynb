##**1.2. Creación de Gráficos Básicos**

```python
import matplotlib.pyplot as plt
```

```python
# Datos para el gráfico de líneas
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 35, 20]

# Crear un gráfico de líneas
plt.plot(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')
plt.show()
```

```python
# Datos para el gráfico de dispersión
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 35, 20]

# Crear un gráfico de dispersión
plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Dispersión')
plt.show()
```

##**1.3. Tipos de Gráficos**

```python
# Gráfico de barras
etiquetas = ['Manzanas', 'Plátanos', 'Naranjas']
valores = [30, 45, 20]
plt.bar(etiquetas, valores)
plt.xlabel('Frutas')
plt.ylabel('Cantidad')
plt.title('Gráfico de Barras')
plt.show()
```

```python
# Histograma
datos = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65]
plt.hist(datos, bins=5)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.show()
```

##**1.4. Combinando Matplotlib con Numpy**

```python
import numpy as np
# Datos para el gráfico
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico de la función seno
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Sin(X)')
plt.title('Gráfico de la función seno')
plt.legend(title="Valores de Seno", loc="upper right")
plt.show()
```

```python
# Datos para el gráfico
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Crear un gráfico de la función coseno
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Cos(X)')
plt.title('Gráfico de la función coseno')
plt.legend(title="Valores de coseno", loc="upper right")
plt.show()
```

```python
# Datos para el gráfico
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.cos(x)

# Crear un gráfico de la función sen(x)*cos(x)
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Sen(X)*Cos(X)')
plt.title('Gráfico de la función sen(x)*cos(x)')
plt.legend(title="Valores de sen(x)*cos(x)", loc="upper right")
plt.show()
```

##**1.6. Creación de gráficos 3D**

```python
# Librería necesaria
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Crear datos para el eje x y z
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
```

```python
# Creamos matrices cuadradas repitiendo los mismos valores de x y y
x, y = np.meshgrid(x, y)
```

```python
# Función para calcular los valores de z
z = np.sin(np.sqrt(x**2 + y**2))
```

```python
# Crear una figura y un eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear el gráfico de superficie 3D
surface = ax.plot_surface(x, y, z, cmap='viridis')

# Agregar etiquetas a los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Agregar título al gráfico
ax.set_title('Gráfico de Superficie 3D')
```

##**2.2. Creación de Gráficos**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos incorporado "tips"
tips = sns.load_dataset("tips")
```

```python
# Crear un gráfico de dispersión con Seaborn
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')
plt.title('Gráfico de Dispersión')
plt.show()
```

```python
# Crear un histograma con Seaborn
sns.histplot(data=tips, x="total_bill", bins=20)
plt.xlabel('Total de la Cuenta')
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.show()
```

##**3.5. Ejercicio: Análisis de ventas del primer semestre del 2023**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv("ventas.csv")
```

```python
# Calcular el promedio de ventas
promedio_ventas = data['Ventas'].mean()
```

```python
print(f"Promedio de Ventas: {promedio_ventas:.2f}") # 183.33
```

```python
# Crear un gráfico de barras con Seaborn y Pandas
sns.barplot(data=data, x='Mes', y='Ventas')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas del primer semestre del 2023')

# Agregar una línea vertical para el promedio
plt.axhline(y=promedio_ventas, color='r', linestyle='--', label='Promedio')
plt.legend()
plt.show()
```

# **5) EJERCICIOS**

##**5.1. Creación del dataframe**

```python
import pandas as pd
import numpy as np

# Generar datos aleatorios
np.random.seed(0)
data = {
    'Nombre': np.random.choice(['Michael', 'Jim', 'Dwight', 'Stanley', 'Oscar'], 10),
    'Edad': np.random.randint(30, 55, 10),
    'Puntuacion': np.random.randint(0, 101, 10)
}

# Crear DataFrame
df = pd.DataFrame(data)
```

##**5.2. Guardar el CSV**

```python
# Guardar DataFrame como CSV
df.to_csv('ejercicios.csv', index=False)

# Mostrar el DataFrame
print("DataFrame:")
print(df)
```

##**5.3. Lee el archivo CSV "datos.csv" en un DataFrame llamado df y muestra las primeras 5 filas.**

```python
# Leer el archivo CSV en un DataFrame llamado df y mostrar las primeras 5 filas
print("Primeras 5 filas del DataFrame:")
print(df.head())
```

##**5.4. Encontrar la edad promedio de las personas en el DataFrame.**

```python
# Encontrar la edad promedio de las personas en el DataFrame
edad_promedio = df['Edad'].mean()
print("Edad promedio:", edad_promedio)
```

##**5.5. Encontrar la persona con la puntuación más alta en el DataFrame.**

```python
# Encontrar la persona con la puntuación más alta en el DataFrame
persona_puntuacion_maxima = df.loc[df['Puntuacion'].idxmax()]
print("Persona con puntuación más alta:")
print(persona_puntuacion_maxima)
```

##**5.6. Filtrar el DataFrame para mostrar solo las personas con puntuación superior a 70.**

```python
# Filtrar el DataFrame para mostrar solo las personas con puntuación superior a 70
personas_puntuacion_alta = df[df['Puntuacion'] > 70]
print("Personas con puntuación superior a 70:")
print(personas_puntuacion_alta)
```

##**5.7. Agregar una nueva columna llamada "Categoria" al DataFrame. Asignar la categoría "Joven" a personas menores de 25 años y "Adulto" a personas de 25 años o más**

```python
# Agregar una nueva columna llamada "Categoria" al DataFrame
# Asignar la categoría "Joven" a personas menores de 25 años y "Adulto" a personas de 25 años o más
df['Categoria'] = np.where(df['Edad'] < 25, 'Joven', 'Adulto') # Hacer cambios de nombres se le conoce como MAPEAR
print("DataFrame con nueva columna 'Categoria':")
print(df)
```

##**5.8. Guardar el DataFrame modificado en un nuevo archivo CSV llamado "datos_modificados.csv".**

```python
# Guardar el DataFrame modificado en un nuevo archivo CSV llamado "datos_modificados.csv"
df.to_csv('datos_modificados.csv', index=False)
print("DataFrame modificado guardado como 'datos_modificados.csv'.")
```

##**5.9. Cambio de etiquetas.**
Esto está chido cuando hacemos algún algoritmo de clasificación C:

```python
# Cambiar las categorías "Joven" y "Adulto" a números (0 y 1 respectivamente)
categorias_numeros = {'Joven': 0, 'Adulto': 1}
df['Categoria'] = df['Categoria'].map(categorias_numeros)
print("DataFrame con categorías cambiadas a números:")
print(df)
```
